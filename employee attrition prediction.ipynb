```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,random_state =0)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
import pickle
pickle.dump(sc,open('transform.save','wb'))
```

```python
x_train.shape
```

```python
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(x_train,y_train)
```

```python
lpred = logistic.predict(x_test)
```

```python
lpred
```

```python
from sklearn.metrics import accuracy_score
laccuracy = accuracy_score(y_test,lpred)
```

```python
laccuracy
```

```python
from sklearn.metrics import confusion_matrix
lcm= confusion_matrix(y_test,lpred)
lcm
```

```python
import sklearn.metrics as metrics
lfpr,ltpr,lthreshold = metrics.roc_curve(y_test,lpred)
lroc_auc = metrics.auc(lfpr,ltpr)
```

```python
lroc_auc
```

```python
plt.title('graph representation on employee attrition')
plt.plot(lfpr,ltpr,color ="blue",label ="Auc_roc curve value=%0.2f"%lroc_auc)
plt.legend()
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel("NoofCompaniesWorked")
plt.ylabel("YearsAtCompany")
```
